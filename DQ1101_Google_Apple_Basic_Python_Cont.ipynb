Below is a list of functions that I can reuse for the analysis

```python
def fileRead(fileName, listFile=True):
    from csv import reader
    readFile = reader(open(fileName, encoding="utf8"))
    
    if listFile:
        return list(readFile)
    else:
        return readFile
    
def dataSample(dataset, start, end):
    print('\n')
    print('Here are some sample data:')
    print('\n')
    for row in dataset[start:end]:
        print(row)
        print('\n') # adds a new (empty) line after each row
    
def dataExplore(dataset, rowcolCount=True, headerInfo=True):            
    if rowcolCount:
        print('\n')
        print('Number of rows:', len(dataset[1:]))
        print('Number of columns:', len(dataset[0]))

    if headerInfo:
        print('\n')
        print('Column headers:')
        index = 0
        for column in dataset[0]:
            print(str(index) + ': ' + column)
            index += 1

def tableFreq(dataset, index):
    freqDict = {}
    
    for row in dataset:
        x = row[index]
        if x in freqDict:
            freqDict[x] += 1
        else:
            freqDict[x] = 1
            
    for key in freqDict:
        print(key + ": " + str(freqDict[key]))

def indexSearch(dataset, searchValue):
    indexValue = 0
    indexDict = {}
    
    for d in dataset:
        indexValue += 1
        if searchValue in indexDict:
            indexDict[indexValue] = searchValue
        
    for key in indexDict:
        print(key + ": " + str(indexDict[key]))
```
